```python
# Histogram for the distribution of carbon emission by vehicle class
from matplotlib import colors
plt.figure(figsize=(12, 6))
N, bins, patches = plt.hist(df['co2'], bins=16, edgecolor='black', alpha=0.7)
fracs = ((N**(1 / 5)) / N.max())
norm = colors.Normalize(fracs.min(), fracs.max())

# Apply colormap to the histogram bars
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)
plt.title('Distribution of CO2 Emission', pad=10)
plt.xlabel('CO2 Emission (g/mile)', labelpad=10)
plt.ylabel('Frequency', labelpad=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Histogram for the ghgScore
plt.figure(figsize=(12, 6))

plt.hist(df['ghgScore'], bins = 8, edgecolor='red', color='lightgreen')
plt.title('Histogram for the Green House Gas Score', pad=10)
plt.xlabel('Green house gas Score (ghgscore)', labelpad=5)
plt.ylabel('Frequency', labelpad= 5)
plt.grid(True, linestyle='--', alpha= 0.7)
plt.show()
```

```python
# Histogram for the Engine displacement
plt.figure(figsize=(16, 12))
plt.hist(df['displ'], bins = 8, edgecolor='black', color='lightgrey')
plt.title('Histogram for the Engine displacement')
plt.xlabel('Engine displacement (displ)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha= 0.7 )
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths= 0.5, center= 0, cbar=False)
plt.title('Correlation Matrix Heatmap of the Fuel Economy dataset')
plt.show()
```

- There is a strong positive correlation between the greenhouse gas global score and the FE score; this means that an increase in the FE score would likely lead to an increase in the GHG score.

- There's also a positive correlation between the number of cylinders and the displacement; an increase in the volume of the cylinders leads to a higher engine displacement.

- There's a negative correlation between carbon dioxide and the greenhouse gas global score. A decrease in the level of carbon dioxide is associated with a decrease in the GHG score.
